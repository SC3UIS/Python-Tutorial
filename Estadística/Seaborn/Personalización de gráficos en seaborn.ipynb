# Personalización de gráficos en seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Cambio de estilo y colores

```python
# sns.set_style()     Modificación del estilo del gráfico
# sns.set_palette()   Modificación de la paleta de colores
# sns.set_context()   Modificación de la escala (paper, talk, notebook, poster)
```

```python
# Importación de conjunto de datos

df = pd.read_csv("Students.csv", index_col=0)
```

```python
sns.set_style("whitegrid") # Modificar el estilo, withegrid= fondo blanco con rejilla. Otros valores son [white, dark, darkgrid]
sns.set_palette("viridis")    # Modifica la paleta de colores para todos los gráficos, en este caso paleta=viridis
sns.set_context("notebook")
```

```python
sns.catplot(x="sex", y="G3", data=df, kind="box")
plt.xlabel("Sexo",fontsize=18)
plt.xticks(fontsize=12)
plt.show()
```

## Agregar títulos y etiquetas

```python
# Las paletas de colores también pueden ser personalizadas a voluntad, definiendo en una lista los colores
sns.set_palette(["#39A7D0", "#21c4bf"])

g = sns.catplot(x="sex", y="G3", data=df, kind="box") # Se guarda el gráfico como una variable
g.fig.suptitle("Diagramas de cajas", y=1.05)          # Se usan los métodos de la variable
plt.xlabel("Sexo",fontsize=18)
plt.xticks(fontsize=12)
plt.show()
```

Seaborn considera diferente método dependiendo del tipo de gráfico, por ejemplo: modificación de ejes. Los gráfico obtenidos con la función catplot o relplot son de tipo facegrid, mientras que pointplot, countplot o boxplot son AxesSubplot 

```python
plt.figure(figsize=(5,5))
g = sns.catplot(x="sex", y="G3", data=df, kind="box") # Se guarda el gráfico como una variable
g.fig.suptitle("Diagramas de cajas", y=1.05)          # Se usan los métodos de la variable
g.set(xlabel="Sexo", ylabel="G3")
plt.xticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(5,5))
g = sns.boxplot(x="sex", y="G3", data=df)          # Se guarda el gráfico como una variable
g.set_title("Diagramas de cajas", y=1.05)          # Se usan los métodos de la variable
g.set(xlabel="Sexo", ylabel="G3")
plt.xticks(fontsize=12)
plt.show()
```
