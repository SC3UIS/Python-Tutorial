###  2. Creación de gráficas<a id='Cap2'></a>

```python
# El primer paso para la creación de las gráficas es crear una figura y los ejes, así:

fig = plt.figure()
ax= plt.axes()
```

```python
# Gráfica de una función seno y coseno

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000) # Variable independiente
y = np.sin(x)
z = np.cos(x)

plt.axis([0,10,-1.5,1.5]) # Fijar el limite de los ejes
ax.plot(x, y)      # Se traza el gráfico sobre los ejes previamente definidos
ax.plot(x, z, color="green", linestyle="-.") # Se hace un cambio en el color y estilo para la función coseno
plt.title("Seno y coseno",fontsize=16)       # Se agrega título al gráfico
plt.show()
```

```python
# Gráfica orientada a objetos

ax = plt.axes()
ax.plot(x, np.sin(x),label="Seno")
ax.plot(x, np.cos(x),label="Coseno", color="green", linestyle="-.")
ax.set(xlim=(0, 10), ylim=(-1.5, 1.5),
       xlabel='x', ylabel='sin(x)',
       title='Seno-Coseno') # ax.set permite configurar los gráficos en una única función
ax.legend();
```

## ¿Cómo ajustar los valores predeterminados?

Matplotlib permite ajustar los valores que vienen por defecto, esto con el fin de crear un gráfico personalizado. A continuación se dará paso a cómo modificar el color, cambiar los ejes, el ancho de la linea, insertar leyendas, entre otros.

### Establecer las marcas (ticks) y etiquetas (tick labels) en los ejes

```python
plt.figure(figsize=(8,6)) # Tamaño de la figura

x = np.linspace(-np.pi, np.pi, 256)    # 256 valores entre -pi y pi
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, color="blue", linewidth=1.3, linestyle="-")
plt.plot(x, z, color="red", linewidth=1.4, linestyle="-.")


# ajustar x,y ticks

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],[r'$-1$', r'$0$', r'$+1$'])

# ajustar límites

plt.xlim(x.min()*1.01, x.max()*1.01)

#Mostrar resultado en la pantalla
plt.show()
```

###  <a id='Cap3'></a>3. Configuración de la gráfica

Una de las primeras configuraciones que se suele realizar es la del tipo y color de las líneas. En el momento que se crea el gráfico "plot" se especifica el color con la paralabra "color" y el tipo de línea con "linestyle"

```python
?plt.plot
```

```python
# Color de las líneas

plt.plot(x, np.sin(x - 0), color='blue')        # nombre del color
plt.plot(x, np.sin(x - 1), color='g')           # código corto de color (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Escala de grises entre 0 y 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # tupla RGB, valores de 0 a 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # colores soportados por HTLM
```

```python
# Tipo de línea

plt.plot(x, x + 4, linestyle='-')  # solida
plt.plot(x, x + 5, linestyle='--') # discontinua
plt.plot(x, x + 6, linestyle='-.') # guion y punto
plt.plot(x, x + 7, linestyle=':'); # punteada

# Si no se especifican los colores matplot lo hace por defecto
```

###  <a id='Cap5'></a>5. Gráficos de dispersión

Otro tipo de gráfico de gran utilidad son los de dispersión. Para la realización de estos se puede acudir a la función scatter o declarar el tipo de línea, que en este caso serían puntos

```python
# Alternativa 1: Función scatter

y = np.sin(np.linspace(0,5,50))
x = np.linspace(0,5,50)

plt.figure("Scatter") # Se genera una figura llamada Scatter
plt.title("Puntos")
plt.scatter(x, y, alpha=0.5) # El argumento alpha determina la transparecia de los puntos
plt.show()
```

```python
# Alternativa 2: Tipo de línea

y = np.sin(np.linspace(0,5,50))

plt.plot(y, "x", label="Normal")
plt.plot(0.5*y, ".", label="Menor")
plt.plot(1.5*y, "2", label="Mayor")
plt.axhline(0, linestyle="--", linewidth=0.5)
plt.legend();
```

Una de las principales diferencia entre el uso de plot y scatter es la posibilidad de controlar de forma individual a cada observación

```python
np.random.RandomState(0) # Fijar semilla para generación de números aleatorios

x = np.random.random(100)
y = np.random.random(100)
colors = np.random.rand(100) # Se define un color por cada valor
size = 1000*np.random.random(100)

plt.scatter(x, y, c=colors, s=size, alpha=0.5, cmap="viridis")
plt.colorbar();
```

### Tipos de gráfico (I)

```python
x = np.arange(100)
y = np.random.rand(100)
plt.plot(x, y, c="black", label="(x,f(x))", linewidth=0.8)
plt.minorticks_on() # Ver marcadores entre los valores de x y y

#Destacar los valores mayores a 0.9

plt.plot(x[y>0.9], y[y>0.9], "o", label="f(x)>0.9")
plt.axhspan(0.9,1,alpha=0.2)
plt.title("Representación de (x,f(x))")
plt.xlabel("Valores de x")
plt.ylabel("Valores de y")
plt.legend();
```

```python
# Se grafica una serie de tiempo, donde las líneas punteadas son las desviaciones

x = np.arange(25)+1
y = np.random.rand(25)*10
y_norm = (y- np.mean(y))/ np.std(y)

plt.xlim(np.min(x)-1,np.max(x)+1)
plt.ylim(np.min(y_norm)-1,np.max(y_norm)+1)

plt.stem(x[y_norm>0],
        y_norm[y_norm>0],
        linefmt="k-.",
        markerfmt="go",
        basefmt="b-")         # Gráfico de tallos. Las letras en los marcadores asignan los colores

plt.stem(x[y_norm<0],
        y_norm[y_norm<0],
        linefmt="k-.",
        markerfmt="ro",
        basefmt="b-")         # Gráfico de tallos. Las letras en los marcadores asignan los colores
plt.show()
```

```python
# Crear polígonos cuando los valores de una serie sean mayores a la otra serie
x = np.arange(25)+1
y_1 = np.random.rand(25)*10
y_2 = np.random.rand(25)*10

plt.plot(x,y_1,color="green", linewidth=1, label="serie 1")
plt.plot(x,y_2,"k-.", linewidth=1, label="serie 2")

plt.fill_between(x, y_1, y_2, where=(y_1>y_2), interpolate=True)
plt.title("Serie de tiempo")
plt.legend();
```

### Gráficos circulares (torta)

La función pie () le permite crear gráficos circulares. Las características opcionales incluyen el etiquetado automático del porcentaje de área, la explosión de una o más cuñas desde el centro del pastel y un efecto de sombra.

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # solo se explota/resalta la segunda rebanada (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')   # garantiza que el pastel se dibuje como un círculo.
plt.show()
```

### Gráficos de barras

```python
n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.axes([0.025,0.025,1,1])
plt.bar(X, +Y1, facecolor='darkcyan', edgecolor='white')
plt.bar(X, -Y2, facecolor='darkblue', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.05, y+0.05, '%.2f' % y, ha='center', va= 'bottom') # Bucle para poner etiquetas a las barras superiores

for x,y in zip(X,Y2):
    plt.text(x+0.05, -y-0.05, '%.2f' % y, ha='center', va= 'top')   # Bucle para poner etiquetas a las barras inferiores

plt.xlim(-.5, n)
plt.xticks([])     # Quitar los marcadores al eje x
plt.ylim(-1.25, +1.25)
plt.yticks([])     # Quitar los marcadores al eje y

plt.show()
```

```python
np.random.seed(1)

y=np.random.normal(0.5,0.4,1000)
y=y[(y>0)&(y<1)]
y.sort()
x=np.arange(len(y))


# Lineal
plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.title("Lineal")

# logarítmica
plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")
plt.title("Logaritmo")

#Simétrica

plt.subplot(2,2,3)
plt.plot(x,y-y.mean())
plt.yscale("symlog",linthreshy=0.01)
plt.title("Simétrica")

#Logística
plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Logit")

# Reducir las etiquetas de los ejes
plt.gca().yaxis.set_minor_formatter(NullFormatter())

plt.subplots_adjust(top=0.92,bottom=0.08,left=0.1,right=0.95,
                   hspace=0.35,wspace=0.35)
```
